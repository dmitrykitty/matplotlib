```python
import numpy as np #najliepiej, by wszystkie dane by≈Çy w np array
```

```python
y = np.array([1, 2, 3]) #x jako indeksy czyli f(0) = 1, f(1) = 2...
plt.plot(y)
plt.show()
```

```python
x = np.array([-1, 0, 1])
y = np.array([1, 0, 1])
plt.plot(x, y) #jezeli przekazywac dwie wartosic - pierwsza x, druga y
plt.plot([-1, 0, 1], [0, 1, 0]) #kilka wykresow na raz. lub dwa razy plot, lub plt.plot(x1, y1, x2, y2, ...)
plt.grid() #w≈ÇƒÖczenie grida
plt.show()
```

```python
plt.plot(x, y, '--c') #trzeci parametr wskazuje na typ linii (--, -., :, -(ba default linia nieprzerywana)), literaz wskazuje na kolor (r, b, w, c, y, g...) 
#lub jako color=b dla wszystkich wykresow (mozna uzywac RGB code lub kod szestastkowy)
plt.grid()
plt.show()
```

```python
x = np.array([-1, 0, 1])
y = np.array([1, 0, 1])
plt.plot([-1, 0, 1], [0, 1, 0],':ro' , x, y, '-cx', markerfacecolor='w') #lub jako marker=... dla wszystkich wykresow. markerfacecolor - kolor wype≈Çnienia
plt.grid() 
plt.show()
```

```python
lines  = plt.plot(x, y)
plt.grid()
plt.setp(lines, linestyle='--', marker='*', markersize=20, markeredgewidth=2, markeredgecolor='r', linewidth=3, markerfacecolor='c')
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)

plt.plot(x,y, 'k')
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.3) #fill between liniƒÖ wykresy x, y i krzywƒÖ y = ... lub y = 0 by default
plt.fill_between(x, y, where=(y > 0), color='c', alpha=0.3)
plt.grid()
plt.show()
```
