```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 4))
fig.set(facecolor='#eee') 

ax = fig.add_subplot()
#ax.set(facecolor='#eee') same as for fig to provide some params
plt.figtext(0.13, 0.9, 'Some text') #some text inside the window

fig.suptitle('Title 1') #title for the window

ax.set_xlabel('Values for x', fontsize=14) #title for x 
ax.set_ylabel('Values for y', fontsize=14) #title for y
#if we have just one x, y pair
#plt.xlabel('Text')
#plt.ylabel('Text')

x = np.arange(-np.pi / 2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.text(-1, 0.75, 'Sinus!!!', color='r', fontsize=14, bbox={'boxstyle':'round', 'facecolor':'#AAAAFF'}) #text inside graph area

ax.annotate('Sinus max value = 1', xy=(np.pi / 2, 1), xytext=(1, 0.5), 
            arrowprops={'facecolor':'gray', 'shrink':0.001})
ax.grid()
```

```python
fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot()
fig.suptitle('Sin + Cos') 
ax.set_xlabel('Ox', fontsize=14) 
ax.set_ylabel('Oy', fontsize=14) 

x = np.arange(-np.pi / 2, np.pi, 0.1)

line1, = ax.plot(x, np.sin(x), '-r') #also label could be added here instead ax.legend - label='text'
line2, = ax.plot(x, np.cos(x), ':b')
line3, = ax.plot(np.array([-1, 0, 1, 2, 3]), np.array([0, 0.5, -0.5, -0.75, 0.75]), 'yo--')

ax.legend((line1, line2, line3), [r'$f(x) = sin(x)$', r'$f(x) = a \cdot b$', r'$f(x) = x \pm \varepsilon$'], loc='upper left', edgecolor='k') #location of the legend, bbox_to_anchor=(1, 0.7) - coordinates if needed
                                    #support of tex notation (simple latex code example)

ax.grid()
```
