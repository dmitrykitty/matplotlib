```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#subplot(nrows, ncols, index)
#draw graphs one by one 

ax1 = plt.subplot(2, 3, 1)
plt.plot(np.random.random(10)*10)

ax2 = plt.subplot(2, 3, 2)
plt.plot(np.random.random(10)*10)

ax3 = plt.subplot(2, 3, 3)
plt.plot(np.random.random(10)*10)

ax4 = plt.subplot(2, 1, 2)
plt.plot(np.random.random(10)*10)

ax1.grid()
ax2.grid()
ax3.grid()

```

```python
f, ax = plt.subplots(2, 2) #(nrow, ncol) -> figure as f and list of coordinates as ax
f.set_size_inches(10, 8) #set the dimencion of the figure
f.set_facecolor('#eee') #ustawienie koloru (jasno szary)
ax[0,0].plot(np.arange(0, 5, 1), '-o')
ax[0,0].grid()
ax[0,1].plot(np.arange(5, 0, -1), '-xr', lw=0.3)
ax[0,1].grid()

fig = plt.figure(figsize=(10, 8)) #set second figure
plt.plot(np.random.random(10) * 10, 'o-')
plt.grid()
```

```python
from matplotlib.gridspec import GridSpec #dla bardziej uniwersalnego rozmieszczenia wykresow

ws = [2, 2, 7] #jaka czesc zajmuje kazda kolumna
hs = [2, 1] #jaka czesc zajmuje kazdy wiersz

fig = plt.figure(figsize=(8, 6))
gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)



ax1 = plt.subplot(gs[0, 0]) #0 row, 0 col
ax1.plot(np.arange(0, 5, 1), '-*c')
ax2 = fig.add_subplot(gs[1, 0:2]) # 1st row, 0,1 col
ax2.plot(np.random.random(10) * 10, 'o-')
ax3 = fig.add_subplot(gs[0:2, 2])
ax3.plot(np.random.random(10)*10, ':xr')
```
