```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2

x = np.linspace(-5, 5, 100)
y = f(x)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
```

```python
#dla edytowania fortmatu liczn na osiach - Formatters
from matplotlib.ticker import NullFormatter

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)

ax.xaxis.set_major_formatter(NullFormatter()) #usuniecie liczb na podziałach 
ax.yaxis.set_major_formatter(NullFormatter()) 
ax.grid()
```

```python
from matplotlib.ticker import FormatStrFormatter


fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)

ax.yaxis.set_major_formatter(FormatStrFormatter("y = %.4f")) #przyjmuje string z formatem
ax.grid()
```

```python
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import MultipleLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)

def formatOy(x, pos):
    return (x if x == 0 else f"[{x}]") if x >= 0 else f"({x})"

ax.xaxis.set_major_formatter(FuncFormatter(formatOy)) #przyjmuje funkcje z dwoma argumentami. Pozwala wskazac format przez zadaną funkcje
ax.xaxis.set_major_locator(MultipleLocator(base=1))
ax.grid()
```

```python
f, ax = plt.subplots(1, 2)
f.set_size_inches(10, 5)

y1 = np.array(y * 10e3)
ax[0].plot(x, y1)
ax[0].grid()

y2 = np.array(y * 10e6)
ax[1].plot(x, y2)
ax[1].grid()
```

```python
from matplotlib.ticker import ScalarFormatter

f, ax = plt.subplots(1, 2)
f.set_size_inches(10, 5)

y1 = np.array(y * 10e3)
ax[0].plot(x, y1)
ax[0].grid()
sf = ScalarFormatter()
sf.set_powerlimits((-3, 3)) #od potegi -3 i 3 będzie skalowanie
ax[0].yaxis.set_major_formatter(sf)

y2 = np.array(y * 10e6)
ax[1].plot(x, y2)
ax[1].grid()
sf = ScalarFormatter()
sf.set_powerlimits((-9, 9)) #od potegi -9 i 9 będzie skalowanie
ax[1].yaxis.set_major_formatter(sf)

#aby takie ustrawienie działało dla wszystkich wykresow 
#matplotlib.rcParams["axes.formatter.limits"] = [-4, 4]
```

```python
from matplotlib.ticker import FixedFormatter
from matplotlib.ticker import FixedLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)

ax.xaxis.set_major_formatter(FixedFormatter(['a', 'b', 'c'])) #pozwala wskazac konkretne wartosci na line grids. Uzywac wraz z FixedLocator
ax.xaxis.set_major_locator(FixedLocator([-3, 0, 3]))
ax.grid()
```
