```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set(xlim=(-5, 30), ylim=(-1, 6)) #ustawienie granic na osiach

#lub osobno dla kazdej osi
ax.set_xlim(0, 15)
ax.set_ylim(ymax=5) #ymin będzie wyliczone automatycznie
```

```python
from matplotlib.ticker import NullLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.grid()
ax.xaxis.set_major_locator(NullLocator()) #xaxis... - horyzontal, yaxis... - vertical 
```

```python
from matplotlib.ticker import MultipleLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set_xlim(xmax=6)
ax.set_ylim(ymax=3)

ax.grid()
ax.xaxis.set_major_locator(MultipleLocator(base=1)) #base * i. gdzie i - nb of grid line
ax.yaxis.set_major_locator(MultipleLocator(base=.5)) #base * i. gdzie i - nb of grid line
```

```python
from matplotlib.ticker import IndexLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi / 2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0.57)) #base * i. gdzie i - nb of grid line. offset - start point + offset (-pi/2 + 0.57)
```

```python
from matplotlib.ticker import FixedLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi / 2, np.pi, 0.1)
ax.plot(x, np.cos(x))
ax.set_xlim(-2.5, 3.5)

ax.grid()
ax.xaxis.set_major_locator(FixedLocator(np.arange(-2, 4, 1))) #ustawia grid line w wyznaczonych punktach
```

```python
from matplotlib.ticker import LogLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi / 2, 3 * np.pi, 0.1)
ax.plot(x, np.cos(x))
ax.set_xlim(-2.5, 10)

ax.grid()
ax.xaxis.set_major_locator(LogLocator(base=2)) #base ^ i. By default base=10
```

```python
from matplotlib.ticker import MaxNLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi / 2, 3 * np.pi, 0.1)
ax.plot(x, np.cos(x))
ax.set_xlim(-2.5, 10)

ax.grid()
ax.xaxis.set_major_locator(MaxNLocator(10)) #rozbiaja na max N linii, ale moze byc i mniej, jezeli wartosci będą równe
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi / 2, 3 * np.pi, 0.1)
ax.plot(x, np.cos(x))
ax.set_xlim(-2.5, 10)

ax.minorticks_on() #dodatkowy minor grid
ax.grid(which='major',lw = 1.2)
ax.grid(which='minor', lw = 0.4)

ax.xaxis.set_major_locator(FixedLocator(np.arange(-2, 10, 1))) #rozbiaja na max N linii, ale moze byc i mniej, jezeli wartosci będą równee
ax.xaxis.set_minor_locator(NullLocator()) #rozbiaja na max N linii, ale moze byc i mniej, jezeli wartosci będą równee
```
